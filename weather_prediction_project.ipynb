```python
#DROPING UNWANTED
dfe.drop(columns=['SNOW','SNWD'],inplace=True)
dfe
```

```python
#CHECKING MISSING VALUES
dfe.isna().sum()
```

```python
#FILLING O VALUES WITH NaN
dfe.loc[dfe.PRCP==0,'PRCP']=np.NaN
dfe.loc[dfe.TMAX==0,'TMAX']=np.NaN
dfe.loc[dfe.TMIN==0,'TMIN']=np.NaN
```

```python
#FILLING THE MISSING WITH MEAN
dfe['PRCP']=dfe['PRCP'].fillna(dfe['PRCP'].mean())
dfe['TMAX']=dfe['TMAX'].fillna(dfe['TMAX'].mean())
dfe['TMIN']=dfe['TMIN'].fillna(dfe['TMIN'].mean())
```

```python
#CHECKING MISSING VALUES
dfe.isna().sum()
```

```python
#DATA TYPE
dfe.dtypes
```

DATA ANALYSING

```python
#ASSIGNING DATAFRAME INTO A VARIABLE
x=dfe
```

```python
#FINDING THE NUMBER OF CLUSTERS
from sklearn.cluster import KMeans
wcss=[] #empty list
for i in range(1,11): #i=clusters
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_) #inertia=for calculating variance
```

```python
wcss
```

```python
#PLOTTING ELBOW POINT
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.xlabel("n_clusters")
plt.ylabel("varience")
plt.title("Elbow point")
```

```python
#SELECTING 4 AS ELBOW POINT
Kmeans=KMeans(n_clusters=4,init='k-means++',random_state=1)
y_pred=Kmeans.fit_predict(x)
y_pred
```

```python
#ASSIGNING CLUSTERS INTO A COLUMN
x['cluster']=y_pred
y_pred
```

```python
x
```

```python
#X AND Y SEPERATION
a=x.iloc[:,:-1].values
b=x.iloc[:,-1].values
```

```python
#TRAINING AND TESTING DATA
from sklearn.model_selection import train_test_split
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=.30,random_state=42)
```

```python
#NORMALISATION
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(a_train)
a_train=scaler.fit_transform(a_train)
a_test=scaler.fit_transform(a_test)
```

IMPLIMENTING DECISION TREE CLASSIFIER ALGORITHM

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(a_train,b_train)
a_pred=model.predict(a_test)
a_pred
```

PERFORMANCE EVALAUTION

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,classification_report
print('score=',accuracy_score(b_test,a_pred))
```

```python
print('result=',confusion_matrix(b_test,a_pred))
```

```python
#CLASSIFICATION REPORT
report=classification_report(b_test,a_pred)
print(report)
```

```python
#CONFUSION MATRIX DISPLAY
label=['0','1','2','3']
result=confusion_matrix(b_test,a_pred)
cmd=ConfusionMatrixDisplay(result,display_labels=label)
cmd.plot()
plt.title('Confusion Matrix')
plt.show()
```
